### Quartile and InterQuartile

```python
data=np.array([4, 5, 5, 5, 6, 6, 6, 6, 7, 7, 7, 8])
data
```

```python
q1 = np.percentile(data,25)
q1
```

```python
q3 = np.percentile(data,75)
q3
```

```python
q4 = np.percentile(data,100)
q4
```

```python
q4-q1
```

## Variance & Std Deviation

```python
data=np.array([4, 5, 5, 5, 6, 6, 6, 6, 7, 7, 7, 8])
data
```

```python
np.var(data)  # variation
```

```python
np.std(data) # std deviation
```
