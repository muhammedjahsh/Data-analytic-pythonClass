### Drop Outliers

```python
df
```

```python
dfNums = df.select_dtypes(include=["number"])
dfNums
```

```python
# 25% quartile
Q1 = dfNums.quantile(0.25)
print(Q1)
# 75%
Q3 = dfNums.quantile(0.75)
print(Q3)
# interquartile range - Q3-Q1
IQR = Q3-Q1
print(IQR)
```

```python
df = dfNums[~( (dfNums < (Q1 - 1.5 * IQR)) |(dfNums > (Q3 + 1.5 * IQR)) ).any(axis=1)]
df
```
