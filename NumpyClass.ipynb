## Aggregation Function

```python
arr2 = np.array([[12,4,7,8],[8,9,19,6]])
arr2
```

```python
print(np.sum(arr2))  # total sum
print(np.sum(arr2,axis=0))  # column wise
print(np.sum(arr2,axis=1))  # row wise
```

```python
# mean
print(np.mean(arr2))
print(np.mean(arr2,axis=0))
print(np.mean(arr2,axis=1))
```

## Concatenation

```python
a = np.array([[2,4],[6,9]])
b  = np.array([[4,5],[12,32]])
print(a)
print(b)
```

```python
print(np.concatenate((a,b),axis=0))  # row wise
```

```python
print(np.concatenate((a,b),axis=1))
```

```python
# stacking funcion -- vstack , hstack
#vsertical stack
a = np.array([[2,4],[6,9]])
b  = np.array([[4,5],[12,32]])

print(np.vstack((a,b)))
```

```python
# horizontal stack
print(np.hstack((a,b)))
```

### Array Splitting

```python
# split - splitting array equal parts
# array_split -- splittin array into unequal parts
# hsplit -- horizontal split (2D array)
# vsplit -- vertical split (2D array)
# dsplit --depth wis split (3D array)
```

```python
# split
arr = np.array([10,20,30,55,76,87,12,8,45])
print(arr)
spliting = np.split(arr,3)
print(spliting)
```

```python
# array _split
arr1 = np.array([10,20,30,55,76,87,12,8,45,56])
parts = np.array_split(arr1,4)
print(parts)
```

```python
arr2 = np.array([[23,89,66],[65,98,8],[12,14,15]])
arr2
```

```python
hsplit = np.hsplit(arr2,3)
hsplit
```

```python
vsplit = np.vsplit(arr2,3)
vsplit
```

```python
# dsplit
df2 = np.array([[[1,2,3,4],[7,6,8,9]]])
df2
```

```python
df2.shape  # main bracket (2 bracket)-block, number of list(rows),number of elements
```

```python
parts3 = np.dsplit(df2,2)
parts3
```
