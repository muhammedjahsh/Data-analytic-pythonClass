```python
class rectangle:
    def __init__(self, length, width):
        self.lg = length
        self.wd = width

    def area(self):
        return self.lg * self.wd

    def perimeter(self):
        return 2 * (self.lg + self.wd)
rect1 = rectangle(40,15)
print(rect1.area())
print(rect1.perimeter())

rect2 = rectangle(60,45)
print(rect2.perimeter())
print(rect2.area())
```
