### Merge

```python
data1 = pd.DataFrame({"ID":[101,102,103],"Name":["A","B","c"]})
data1
```

```python
data2 = pd.DataFrame({"empID":[101,102,104],"Salary":[5000,6000,7000]})
data2
```

```python
pd.merge(data1,data2,left_on="ID",right_on="empID")    # by default inner merge/innner join -> common  in both tables
```

```python
# left join/merge
pd.merge(data1,data2,on="ID",how='left') 
```

```python
# right join
pd.merge(data1,data2,on="ID",how='right') 
```

```python
# full outer join - union
pd.merge(data1,data2,on="ID",how='outer') 
```

```python
pd.merge(data1,data2,how='cross',)
```
