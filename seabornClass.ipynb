```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# loading inbulit dataset
data = sns.load_dataset("tips")
data
```

```python
data2 = sns.load_dataset("fmri")
data2
```

## ScatterPlot

```python
sns.scatterplot(data=data,x="total_bill",y="tip",hue="sex")  # hue - color variation among data points
plt.title("Tip by Total Bill correlation") 
plt.show()
```

```python
sns.scatterplot(data=data,x="total_bill",y="tip",style="smoker",hue="sex") 
```

```python
sns.scatterplot(data=data,x="total_bill",y="tip",sizes=(20,200),size="size") 
```

## Line Plot

```python
data
```

```python
sns.lineplot(data=data,x="day",y="total_bill",hue="sex")
plt.title("Day by Total bill")
plt.show()
```

```python
sns.lineplot(data=data,x="day",y="total_bill",style="sex",markers=True,linewidth=3,ci=None)
plt.title("Day by Total bill")
plt.show()
```

## Relational Plots

```python
sns.relplot(data=data,x="tip",y="total_bill")
```

```python
sns.relplot(data=data,x="day",y="total_bill",kind='line')
```

```python
#multiple subplots using relplots
sns.relplot(data=data,x="tip",y="total_bill",row="sex",col="smoker",kind='scatter',hue="time")
plt.title("Tips by Total bills")
plt.show()
```

## Sub Plots

```python
data
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,6))  # rows,columns
sns.scatterplot(data=data,x="total_bill",y="tip",ax=axs[0])  # ax - axes
axs[0].set_title("Scatter plot")
sns.lineplot(data=data,x='day',y='total_bill',ax=axs[1])
axs[1].set_title("Line plot")
plt.tight_layout()
plt.show()
```

```python
# multple rows and columns
_,axs = plt.subplots(2,2,figsize=(18,12))

sns.scatterplot(data=data,x="total_bill",y="tip",ax=axs[0,0])
axs[0,0].set_title("Tip by total bill")
sns.lineplot(data=data,x="day",y="total_bill",ax=axs[0,1])

sns.lineplot(data=data,x="day",y="tip",ax=axs[1,0])

sns.scatterplot(data=data,x="total_bill",y="size",ax=axs[1,1])
```

## Categorical Plotting

### CountPlot

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(data=df,x='day',hue='sex')   # for distribution
plt.title("Number of customers by Gender")
plt.show()
```

```python
sns.countplot(data=df,x='day',order=["Sat","Sun","Thur","Fri"],saturation=0.5)   # for distribution
plt.title("Number of customers by Gender")
plt.show()
```

### Box Plot

```python

sns.boxplot(data=df,x='day',y='tip',linecolor="red",linewidth=3)   # detect outliers and find distribution
plt.title("Tip distribution by day")
plt.show()
```

### ViolinPlot

```python
sns.violinplot(data=df,x='day',y='tip')
plt.title("Tip distribution by day")
plt.show()
```

## Histogram

```python
sns.histplot(data=df,x="tip",bins=3,kde=True)
plt.title("number of customers by tip")
plt.show()
```

## Joint Plot

```python
sns.jointplot(data=df,x="total_bill",y='tip',kind='hex') # combination histogram and scatter plot
plt.title("joint plot")
plt.show()
```

## Pairplot

```python
sns.pairplot(data=df,hue='sex')
plt.tight_layout()
plt.show()
```

## Regression Plot

```python
sns.regplot(data=df,x='total_bill',y='tip')
```
