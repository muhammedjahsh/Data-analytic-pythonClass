### Plot / Line Chart

```python


Name = ["A","B","C","D"]
Mark = [45,36,78,66]
plt.figure(figsize=(12,7))
plt.plot(Name,Mark,marker="D",linestyle="-.",color="Black",linewidth=3,markerfacecolor="Black",markeredgecolor="Blue")
plt.title("Mark Distribution",fontdict={'fontsize':15,'color':'Blue','weight':'bold','style':'italic'},loc='right')   # main title
# weight = light,normal,bold,heavy
# style = normal,italic,oblique
# loc = center(by default),left, right
plt.xlabel("Students Name")   # x-axis title
plt.ylabel("Marks")     # y-axis title
plt.show()   # remove unwanted texts
```

```python
# application
import pandas as pd
```

```python
data = pd.read_excel("campaign (1).xlsx")
data
```

```python
data.dtypes
```

```python
# creating new column
data["Month"] = data["Date"].dt.strftime("%B")
data
```

```python
grouped_data = data.groupby("Month")["Clicks"].sum()
grouped_data
```

```python

plt.plot(grouped_data.index,grouped_data.values,marker = "o",linestyle="-.",color="Black",linewidth=3,markerfacecolor="Black",markeredgecolor="Blue")
plt.title("Monthly Clicks Distribution")
plt.xlabel("Month Name")
plt.ylabel("Clicks")
for x,y in zip(grouped_data.index,grouped_data.values):
    plt.text(x,y+25,str(y))
    
plt.show()
```

## Sub Plots

```python
Name = ["hari","David","Jhon","carls",'Sam']

phy_Mark = [14,16,32,91,56]
che_mark = [63,36,65,56,78]
maths_mark = [15,32,56,96,47]
social_mark = [45,65,32,26,86]

plt.subplot(2,2,1) # rows,columns, position
plt.plot(Name,phy_Mark)
plt.title("Physics Mark")

plt.subplot(2,2,2)
plt.plot(Name,che_mark)
plt.title("Chemistry Mark")

plt.subplot(2,2,3)
plt.plot(Name,maths_mark)
plt.title("Maths Mark")

plt.subplot(2,2,4)
plt.plot(Name,social_mark)
plt.title("Social Mark")

plt.tight_layout()
plt.show()
```

## Scatter Plot

X axis - measures

Y axis - measures

--- uses/purpose -
1) Correlation
2) to detect outliers - check far away datapoint
3) finding patterns or trend
4) distribution of data

```python
phy_Mark = [14,16,32,91,56]
che_mark = [15,36,65,56,78]
plt.scatter(phy_Mark,che_mark,color="red",marker="s",s=150,edgecolors='Black')
plt.xlabel("physics mark")
plt.ylabel("chemistry mark")
plt.grid()  # grid lines
plt.show()
```

```python
# color gradient
phy_Mark = [14,16,32,91,56]
che_mark = [15,36,65,56,78]
plt.scatter(phy_Mark,che_mark,c=che_mark,cmap='Blues',s=200)
plt.colorbar(label="Chemistry mark")
plt.xlabel("physics mark")
plt.ylabel("chemistry mark")
```

```python
# size variation
phy_Mark = [14,16,32,91,56]
che_mark = [15,36,65,56,78]
size = [i*5 for i in phy_Mark]

plt.scatter(phy_Mark,che_mark,s=size)
plt.xlabel("physics mark")
plt.ylabel("chemistry mark")
```

```python
import numpy as np
```

```python
# multple bar plots
Name = ["A","B","C","D"]
phy_mark = [45,36,78,66]
che_mark = [45,63,36,56]
x= np.arange(1,len(Name)+1)  # 1,2,3,4

wid = 0.4
plt.bar(x,phy_mark,width=wid,label="Physics Mark")
plt.bar(x-wid,che_mark,width=wid,label= "Chemistry Mark")
plt.xticks(x,labels=Name)
plt.legend()
plt.show()
```

## Pie Chart

Legend - Dimensions

values - measures

-- maximum 5 or 6 slices



```python
Name = ["A","B","C","D"]
phy_mark = [45,36,78,66]
plt.figure(figsize=(6,6))
cl = ['yellow','red','violet','blue']
ex = [0,0,0,0.2]
plt.pie(phy_mark,labels=Name,colors=cl,explode=ex,autopct='%i%%',shadow=True,startangle=180)   # i indicated intger ,  f indicate float
plt.title("Total Marks Distribution")
plt.show()
```

## Stack Plots

```python
Name = ["A","B","C","D"]
phy_mark = [45,36,78,66]
cheMark = [49,65,63,48]
mathsMark = [96,36,46,68]

plt.stackplot(Name,phy_mark,cheMark,mathsMark,labels=["Physics mark","chemistry mark","maths mark"],colors=["red","green","yellow"])
plt.legend(loc="upper center")
plt.xlabel("Names")
plt.title("marks distribution")
plt.show()
```
